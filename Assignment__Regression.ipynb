## **Creating Feature Set(X) and Outcome Variable(Y)**

```python
Q1 = df.price.quantile(0.25)
Q3 = df.price.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
df[(df.price<lower_limit)|(df.price>upper_limit)]
```

```python
df_no_outlier = df[(df.price>lower_limit)&(df.price<upper_limit)]
df_no_outlier
```

```python
df.shape
```

```python
df_no_outlier.shape
```

```python
df['zscore'] = ( df_no_outlier.price - df_no_outlier.price.mean() ) / df_no_outlier.price.std()
df.zscore
```

```python
df[df['zscore']>3]
```

```python
df[df['zscore']<-3]
```

```python
df[(df.zscore<-3) | (df.zscore>3)]
```

```python
df_no_outliers = df[(df.zscore>-3) & (df.zscore<3)]

```
